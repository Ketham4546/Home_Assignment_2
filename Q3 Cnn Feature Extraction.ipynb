```python
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Task 1: Edge Detection Using Sobel Filter
def edge_detection(image_path):
    # Load grayscale image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply Sobel filter in x and y directions
    sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

    # Display images
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
    plt.subplot(1, 3, 2), plt.imshow(sobel_x, cmap='gray'), plt.title('Sobel-X')
    plt.subplot(1, 3, 3), plt.imshow(sobel_y, cmap='gray'), plt.title('Sobel-Y')
    plt.show()

# Task 2: Max Pooling and Average Pooling
def pooling_operations():
    # Create a random 4x4 matrix
    input_matrix = np.random.rand(4, 4).astype(np.float32)
    input_tensor = tf.constant(input_matrix.reshape(1, 4, 4, 1))

    # Apply Max Pooling (2x2)
    max_pooled = tf.nn.max_pool2d(input_tensor, ksize=2, strides=2, padding='VALID')

    # Apply Average Pooling (2x2)
    avg_pooled = tf.nn.avg_pool2d(input_tensor, ksize=2, strides=2, padding='VALID')

    # Print results
    print("Original Matrix:\n", input_matrix)
    print("\nMax Pooled Matrix:\n", max_pooled.numpy().squeeze())
    print("\nAverage Pooled Matrix:\n", avg_pooled.numpy().squeeze())

# Execute functions
if __name__ == "__main__":
    image_path = '/content/Screenshot_2024-02-09-08-59-00-67_1c337646f29875672b5a61192b9010f9~2.jpg'  # Change this to the actual image path
    edge_detection(image_path)
    pooling_operations()
```
